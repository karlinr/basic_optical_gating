```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib as mpl
mpl.get_configdir()
plt.style.use('seaborn-talk')
```

```python

def linear_function(x, a, b):
    return a * x + b

plt.figure(figsize = (10, 7))
xs = np.linspace(-2, 2, 5)
plt.plot(xs, linear_function(xs, 1, 0), lw = 2, solid_capstyle = "round", c = "black", ls = "--")
xs = np.linspace(-1, 1, 3)
plt.scatter(xs, linear_function(xs, 1, 0), zorder=10000, s = 120, c = "tab:blue")
plt.scatter([-0.4], linear_function(-0.4, 1, 0), zorder=10000, s = 240, c = "tab:red")
plt.grid(False)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize = (10, 7))
xs = np.linspace(-2.5, 2.5, 1000)
plt.plot(xs, np.abs(linear_function(xs, 1, 0) - linear_function(0.4, 1, 0)), lw = 2, c = "black", ls = "--")
xs = np.linspace(-2, 2, 5)
plt.scatter(xs, np.abs(linear_function(xs, 1, 0) - linear_function(0.4, 1, 0)), c = "tab:blue", zorder = 1000)
plt.scatter([0.4], linear_function(0, 1, 0), zorder=10000, s = 240, c = "tab:red")
plt.grid(False)
plt.axis("off")
plt.show()
```

```python
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

plt.figure(figsize = (14, 10))
xs = np.linspace(-2, 2, 1000)
plt.plot(xs, quadratic_function(xs, 1, 4, 0), lw = 2, solid_capstyle = "round", c = "black", ls = "--")
xs = np.linspace(-1, 1, 3)
plt.scatter(xs, quadratic_function(xs, 1, 4, 0), zorder=10000, s = 120, c = "tab:red")
plt.scatter([-0.5], quadratic_function(-0.5, 1, 4, 0), zorder=10000, s = 240, c = "tab:green")
plt.grid(False)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize = (10, 7))
xs = np.linspace(-2.5, 2.5, 1000)
plt.plot(xs, np.abs(quadratic_function(xs, 1, 4, 0) - quadratic_function(0.4, 1, 4, 0)), lw = 2, c = "black", ls = "--")
xs = np.linspace(-2, 2, 5)
plt.scatter(xs, np.abs(quadratic_function(xs, 1, 4, 0) - quadratic_function(0.4, 1, 4, 0)), c = "tab:blue", zorder = 1000)
plt.scatter([0.4], quadratic_function(0, 1, 4, 0), zorder=10000, s = 240, c = "tab:red")
plt.grid(False)
plt.axis("off")
plt.show()
```
