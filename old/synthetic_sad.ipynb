```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def superdupergaussian(x, A, x_0, sigma, s, c, d):
    return -A * np.exp(-((x - x_0)**2 / (2 * sigma**2))**s) + c + d * x

xs = np.linspace(-3, 3, 40)

plt.scatter(xs, superdupergaussian(xs, 1, 0, 0.3, 0.4, 7.8, 0.0))
plt.show()
```
