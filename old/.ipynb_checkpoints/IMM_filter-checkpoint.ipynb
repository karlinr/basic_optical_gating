```python
from classes.basic_optical_gating import BasicOpticalGating as BOG, BasicOpticalGatingPlotter as BOGP, Logger
from classes.drawer import Drawer
from classes.kalman_filter import KalmanFilter, IMM as KF, IMM
```

```python
import numpy as np
import matplotlib.pyplot as plt
phase_min = 0
phase_max = 1000
phase_per_frame = 1
phases = np.arange(phase_min, phase_max / 2, phase_per_frame)
phases = np.concatenate((phases, phase_max / 2 + 2 * phases))
plt.plot(phases)
plt.show()
```

```python
v = Drawer(30, 39.1, (256, 256))
v.generate_reference_sequence()
v.generate_sequence()
og = BOG()
og.logger.set_quiet()
og.set_reference_period(v.reference_period)
og.set_reference_sequence(v.reference_sequence)
og.set_sequence(v.sequence)
og.set_reference_framerates(v.reference_phases)
og.run(bias_correct = True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 1
            
# This is a constant acceleration filter
x = np.array([og.unwrapped_phases[0], og.unwrapped_phases[1] - og.unwrapped_phases[0], 0.1])
P = np.diag([0.1, 0.1, 1.])
q = 0.005
Q = np.array([[dt**5 / 20, dt**4 / 8, dt**3 / 6], 
              [dt**4 / 8, dt**3 / 3, dt**2 / 2], 
              [dt**3 / 6, dt**2 / 2, dt]]) * q
R = 1
F = np.array([[1, dt, dt**2 / 2], 
              [0, 1, dt], 
              [0, 0, 1]])
H = np.array([[1, 0, 0]])

# Create filters for our IMM and KF
kf_ca_imm = KalmanFilter(dt, x, P, Q, R, F, H)
            
# This is a constant velocity filter
x = np.array([og.unwrapped_phases[0], og.unwrapped_phases[1] - og.unwrapped_phases[0], 0.])
P = np.diag([0.1, 0.1, 0.1])
q = 0.1
Q = np.array([[dt**3 / 3, dt**2 / 2, 0], 
              [dt**2 / 2, dt, 0], 
              [0, 0, 0]]) * q
R = 1
F = np.array([[1, dt, 0], 
              [0, 1, 0], 
              [0, 0, 0]])
H = np.array([[1, 0, 0]])

# Create filters for our IMM and KF
kf_cv_imm = KalmanFilter(dt, x, P, Q, R, F, H)


# Uses 3 filters: a constant acceleration filter, a constant velocity filter and a steady state filter.
# Add filter to our bank
filters = [kf_ca_imm, kf_cv_imm]

# Probabilities of filters
mu = np.array([0.5, 0.5])
M = np.array([[.86, .14],
              [.1, .9]])

# Initialise interacting multiple model filter
imm = IMM(filters, mu, M)
```

```python
# IMM arrays
kf_imm_unwrapped_phases = []
kf_imm_prob_ca = []
kf_imm_prob_cv = []
kf_imm_prob_ss = []
kf_imm_error = []

#
kf_imm_as = []
kf_imm_vs = []
kf_imm_xs = []

for i in range(len(og.unwrapped_phases)):           
    # Run our IMM filter
    imm.predict()
    imm.update(og.unwrapped_phases[i])
    
    # Save results to an array
    kf_imm_unwrapped_phases.append(imm.x[0])
    kf_imm_prob_ca.append(imm.cbar[0])
    kf_imm_prob_cv.append(imm.cbar[1])
    kf_imm_error.append(imm.P[0,0])

    kf_imm_xs.append(imm.x[0])
    kf_imm_vs.append(imm.x[1])

kf_imm_prob_ca = np.array(kf_imm_prob_ca)
kf_imm_prob_cv = np.array(kf_imm_prob_cv)
```

```python
plt.figure(figsize = (12, 6))
plt.plot(kf_imm_prob_ca, label = "Constant acceleration")
plt.plot(kf_imm_prob_cv, label = "Constant velocity")
plt.legend()
plt.show()

plt.figure(figsize = (12,6))
plt.plot(og.unwrapped_phases - kf_imm_xs)
plt.show()
plt.figure(figsize = (12,6))
plt.plot(kf_imm_vs)
plt.show()
```
